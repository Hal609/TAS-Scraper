```python
def parseInputs(inputs_string):
   inputs = {"p1": [], "p2": [], "p3": [], "p4": []}
   count = 0
   for line in inputs_string.split("\n"):
      inString = "".join([str(int(char != ".")) for char in line[3:11]])
      if len(inString) > 0: inputs["p1"].append(int(inString, 2))
      # print(count, line)
      # if count > 400: break
      # else: count += 1

      inString = "".join([str(int(char != ".")) for char in line[12:20]])
      if len(inString) > 0: inputs["p2"].append(int(inString, 2))

      inString = "".join([str(int(char != ".")) for char in line[21:29]])
      if len(inString) > 0: inputs["p3"].append(int(inString, 2))

      inString = "".join([str(int(char != ".")) for char in line[30:38]])
      if len(inString) > 0: inputs["p4"].append(int(inString, 2))
   return inputs


def readfm2(replayPath) -> tuple[dict, dict]:
   with open(replayPath, "r") as fm2:
      data = fm2.read()
      head = data[:data.find("|0|")]
      inputs_str = data[data.find("|0|"):]
      inputs = parseInputs(inputs_str)

      headDict = {
         stat.split(" ")[0] : " ".join(stat.split(" ")[1:])
         for stat in head.split("\n")
         if stat.strip()
      }

   return inputs, headDict
```

```python
def in2str(inputNum):
   nesIn = f"{inputNum:08b}"
   out = ""
   for i, char in enumerate(posInputs):
      out += char if nesIn[i] == "1" else "-"
   return out
```
