```python
import requests
import pandas as pd

BASE_URL = 'https://tasvideos.org/api/v1'

# 1
# Fetch the list of publications
def get_table(table_name, page_num):
    response = requests.get(f"{BASE_URL}/{table_name}?Systems=NES&PageSize=100&CurrentPage={page_num}")
    response.raise_for_status()  # Raises an error if the request failed
    return response.json()

def get_pubs(page_num):
    response = requests.get(f"{BASE_URL}/publications?Systems=NES&ShowObsoleted=false&PageSize=100&CurrentPage={page_num}")
    response.raise_for_status()  # Raises an error if the request failed
    return response.json()

def get_details(id, table_name):
    response = requests.get(f"{BASE_URL}/{table_name}/{id}?Systems=NES")
    response.raise_for_status()
    return response.json()


# print(f"Total publications: {len(publications)}")
# # print("First publication:", publications[0])

# movie_detail = get_publication_details(1)  # Get details for publication with ID = 1
# print("Publication #1 details:", movie_detail)
```

```python
cols = ['id', 'title', 'branch', 'goal', 'gameGoalId', 'emulatorVersion',
       'class', 'systemCode', 'submissionId', 'gameId', 'gameVersionId',
       'obsoletedById', 'frames', 'rerecordCount', 'systemFrameRate',
       'movieFileName', 'additionalAuthors', 'createTimestamp', 'authors',
       'tags', 'flags', 'urls', 'filePaths']
pubs_df = pd.DataFrame(columns=cols)
for i in range(1, 10):
   page = pd.json_normalize(get_pubs(i))
   pubs_df = pd.concat([pubs_df, pd.json_normalize(get_table("publications", i))])
   print(f"Page {i}/9 fetched")
pubs_df
```

```python
unique_pubs_df = pubs_df.drop_duplicates(subset=['id']).reset_index(drop=True)
unique_pubs_df
```

```python
unique_pubs_df.to_csv("./pubs.csv", index=False)
```
